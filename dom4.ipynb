```python
Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pylab

from scipy.misc import derivative
from scipy.misc import electrocardiogram
from scipy.optimize import fmin, brute
from scipy.signal import find_peaks
```

```python
#f(x) = x^5 + 4sin(2x) + cos(3x+3)
```

```python
def f(x):
    return x**5 + 4 * np.sin(2 * x) + np.cos(3*x + 3)
```

```python
x = np.linspace(-10, 10)
fx = f(x)
f1x = [derivative(f, xx, n=1) for xx in x]
```

```python
# значение 1-ой
derivative(f, 1.0, n=1)
```

```python
# значение 2-ой
derivative(f, 1.0, n=2)
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```

```python
Уровень 1
Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]
```

```python
def f(x):
    return (np.sin(2 * x + 1))**5
```

```python
x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, n=1) for xx in x]
f2x = [derivative(f, xx, n=2) for xx in x]
```

```python
#Постройте график первой и второй производной функции
plt.plot(x, f1x, label='derivative')
plt.plot(x, f2x, label='second derivative')
plt.legend()
plt.show()
```
