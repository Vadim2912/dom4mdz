```python
def f(x):
    return x**5 + 4 * np.sin(2 * x) + np.cos(3*x + 3)
```

```python
x = np.linspace(-10, 10)
fx = f(x)
f1x = [derivative(f, xx, n=1) for xx in x]
```

```python
derivative(f, 1.0, n=1)
```

```python
derivative(f, 1.0, n=2)
```

```python

plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```

```python
def f(x):
    return (np.sin(2 * x + 1))**5
```

```python
x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, n=1) for xx in x]
f2x = [derivative(f, xx, n=2) for xx in x]
```

```python

plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.plot(x, f1x, label='second derivative')
plt.legend()
plt.show()
```
